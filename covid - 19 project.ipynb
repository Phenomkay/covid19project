```python
df = df[["iso_code","continent","location","date","total_cases","total_deaths","total_vaccinations"]]
df   
```

```python
# show countries represented
df["location"].unique()
```

```python
exclude_countries = ["Africa","World","Europe", 'Upper middle income', "Western Sahara", 'Low income', 'Lower middle income',
                     "High income", 'Reunion', "Asia", 'South America',"Oceania","North America","European Union"]
```

```python
# removing the excluded countries
df = df[~ df["location"].isin(exclude_countries)]
df["location"].unique()
```

```python
df[df["location"] == "Nigeria"]["total_cases"].sum()
```

```python
# sort by recent to oldest
data = df.sort_values("date", ascending=False).reset_index()
data
```

```python
# show total cases and total deaths by countries
cpc = df.groupby(["location"])[["total_cases","total_deaths"]].sum().sort_values("total_cases",ascending=False).reset_index()
cpc
```

```python
cpc = round(df.groupby(["location"])[["total_cases","total_deaths"]].sum().sort_values("total_cases",ascending=False).reset_index())
cpc
```

```python
# converting the date column to datetime
df["date"] = pd.to_datetime(df.date)

# change date format
df["date"] = df["date"].dt.strftime('%y/%m/%d')
df
```

```python
# set the date as index
df = df.set_index("date")
df
```

```python
# convert your data to pivot table
data = pd.pivot_table(df,values="total_cases",index=["location"],columns="date").reset_index()
data = data.fillna(0)
data
```

```python
# bring in the flags
flags = pd.read_csv("https://raw.githubusercontent.com/waleolajumoke/flags/main/flags_world.csv")
flags
```

```python
# combining the two data so each country have their corresponding flags
final = pd.merge(data, flags, left_on="location", right_on="country")
final.head()
```

```python
del final["country"]
img = final["Image URL"]
del final["Image URL"]
final.insert(1, 'Flags', img)
final.head()
```

```python
final.to_csv("covidresult.csv")
```
